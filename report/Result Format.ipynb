```python
import pandas as pd
```

```python
df = pd.read_csv("../_output/experiments_result.csv", index_col=False).drop("Unnamed: 0", axis=1)
df["embeddings_type"][df["embeddings_type"] == "bert_small"] = "bert_small (256)"
df["embeddings_type"][df["embeddings_type"] == "bert_tiny"] = "bert_tiny (128)"
```

```python
df.head(10).sort_values(["name"], ascending=False)
```

### Comparing F1-scores of all models

```python
df["exp"] = df["embeddings_type"] + "_" + df["data"]
df2 = df[~df.is_train].sort_values(["data"], ascending=False)
result = df2.pivot(index=["embeddings_type", "data"], columns="name", values=["f1_score"])
```

```python
columns_ordering = [("f1_score","NaiveBayes"), ("f1_score","SVC"), ("f1_score","MLP")]
```

```python
tmp = result.reset_index()
embs_ordering = ["bert_small (256)", "bert_tiny (128)", "w2v", "tfidf"]
tmp["sortingval"] = tmp.embeddings_type.apply(lambda x: embs_ordering.index(x))
tmp = tmp.sort_values(["sortingval", "data"], ascending=False)
tmp.set_index(["embeddings_type", "data"]).drop("sortingval", axis=1)[columns_ordering]
```
