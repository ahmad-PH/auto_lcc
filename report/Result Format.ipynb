```python
df = pd.read_csv("experiments_result_full.csv", index_col=False).drop("Unnamed: 0", axis=1)
df["embeddings_type"][df["embeddings_type"] == "bert_large"] = "bert_large (768)"
df["embeddings_type"][df["embeddings_type"] == "bert_small"] = "bert_small (256)"
df["embeddings_type"][df["embeddings_type"] == "bert_tiny"] = "bert_tiny (128)"
df["accuracy"] = (df["accuracuy"].fillna(0) + df["accuracy"].fillna(0))
#df = df[df["data"]=="both"]
```

### Comparing F1-scores of all models

```python
metric = "precision"

def get_report(metric):
    df["exp"] = df["embeddings_type"] + "_" + df["data"]
    df2 = df[~df.is_train].sort_values(["data"], ascending=False)
    result = df2.pivot(index=["embeddings_type", "data"], columns="name", values=[metric])

    columns_ordering = [(metric,"NaiveBayes"), (metric,"SVC"), (metric,"MLP")]

    tmp = result.reset_index()
    embs_ordering = ["bert_large (768)", "bert_small (256)", "bert_tiny (128)", "w2v", "tfidf"]
    tmp["sortingval"] = tmp.embeddings_type.apply(lambda x: embs_ordering.index(x))
    tmp = tmp.sort_values(["sortingval", "data"], ascending=False)
    return tmp.set_index(["embeddings_type", "data"]).drop("sortingval", axis=1)[columns_ordering]
```

```python
get_report("accuracy")
```

```python
get_report("precision")
```

```python
get_report("recall")
```

```python
get_report("f1_score")
```
